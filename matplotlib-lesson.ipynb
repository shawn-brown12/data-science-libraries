# Matplotlib lesson notebook

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

## Line charts

plt.plot(x values first, y values second)   
Limits - plt.xlim, plt.ylim  
Color - c argument  
Alpha - transparency  
Multiple plots  

```python
x = np.array(range(1, 11))

x
```

```python
y = x ** 2

y
```

```python
plt.plot(y)
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y)
plt.xlim(5, 10)
```

```python
plt.plot(x, y)
plt.xlim(5, 10)
plt.ylim(60, 100)
```

```python
plt.plot(x, y, c='red')
plt.xlim(5, 10)
plt.ylim(60, 100)
```

```python
z = x ** 2 - 4 

plt.plot(x, y, c='oldlace')
plt.plot(x, z, c='firebrick', alpha=.8)
plt.xlim(4, 8)
```

## Scatter charts

Size - s argument

```python
plt.scatter(x, y, s=5000000)
plt.scatter(x, z)
plt.plot(x, y, c='red')
```

## Labels, Titles, and Axes

plt.title  
plt.xlabel  
plt.ylabel  

Write LaTeX symbols by wrapping text with dollar signs ($)

```python
plt.scatter(x, y, s=600)
plt.scatter(x, z)
plt.title('My beautiful curved line of $x^2$')
plt.xlabel('My x values')
plt.ylabel('My y labels')
```

## Custom Ticks

plt.xticks  
plt.yticks  

```python
plt.scatter(x, y, s=600)
plt.scatter(x, z)
plt.xticks(range(1, 11, 2))
plt.yticks([33, 45, 77])
```

## Custom Line Types

Line style - ls argument (accepts ":" and "- -")

```python
plt.plot(x, y, ls=':')
plt.plot(x, z, ls='--')
```

## Rotating text and annotations

Rotation - rotation argument

```python
plt.plot(x, y, ls=':')
plt.plot(x, z, ls='--')
plt.xticks(rotation=340)
```

## Adding text

plt.text - x and y coordinates, text, fontsize, color

plt.annotate - text, pointer coordinates, text coordinates, arrow properties

```python
plt.plot(x, y, ls=':')
plt.plot(x, z, ls='--')
plt.text(2, 10, 'hi', fontsize=32, c='firebrick')
plt.annotate('general kenobi', (7, 75))
plt.annotate(text='hello there', xy=(7, 75), xytext=(2, 60), arrowprops={'width':7})
plt.legend([1, 2])
```

## Figure size and legends

plt.figure - figsize argument to determine size of plot  
plt.legend - loc argument for location

```python
plt.figure(figsize=(5, 5))
plt.plot(x, y, ls=':')
plt.plot(x, z, ls='--')
plt.legend()
```

## Saving figures

plt.savefig - input the filename, default .png filetype

## Subplots

plt.subplot - number of rows, number of columns, selected figure  
plt.subplot - shorthand notation  
plt.suptitle - title for collection of subplots

```python
plt.subplot(1, 2, 1)
plt.plot(x, y)

plt.subplot(1, 2, 2)
plt.plot(x, z)
```

```python
plt.subplot(121)
plt.plot(x, y)
plt.title('y tho', fontsize=8)

plt.subplot(122)
plt.plot(x, z)
plt.title('smol title', fontsize=8)

plt.suptitle('Comparision between two graphs.')
```

plt.hist - values, bins argument align property

```python
q = np.array([1, 1, 2, 1, 4, 2, 4, 5, 6, 3, 4, 2, 3, 3, 5, 3])
```

```python
plt.hist(q)
```

```python
plt.hist(q, bins=[.5, 1.5, 2.5, 3.5, 4.5, 5.5, 6.5], align='mid', ec='firebrick')
```
