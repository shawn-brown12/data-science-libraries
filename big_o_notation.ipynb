```python
import matplotlib.pyplot as plt
import numpy as np
```

- Title your chart "Big O Notation"
- Label your x axis "Elements"
- Label your y axis "Operations"
- Label your curves or make a legend for the curves
- Use LaTex notation where possible

Curves to graph

- a = 0 n + 1  and label the curve "O(1)"
- b = log(n)  and label the curve "O(log n)"
- c = n  and label the curve "O(n)"
- d = n âˆ— log(n)  and label it "O(n log n)"
- e = n**2  and label it "O(n^2)"
- f = 2**n  and label it "O(2^n)"
- g = n !  and label it "O(n!)"
- h = n**n  and label it "O(n^n)"

```python
x = np.array(list(range(1, 11, 2)))

plt.figure(figsize = (12, 6))

a = [0 * x + 1 for x in x]
b = [np.log(x) for x in x]
c = [x for x in x]
d = [x * np.log(x) for x in x]
e = [x ** 2 for x in x]
f = [2 ** x for x in x]
g = [np.math.factorial(x) for x in x]
h = [x ** x for x in x]

plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operation Results')
plt.plot(x, a, c='skyblue', label='$O(1)$', ls='--')
plt.plot(x, b, c='blue', label='$O(!x)$', ls=':')
plt.plot(x, c, c='green', label='$O(x)$', ls='--')
plt.plot(x, d, c='red', label='$O(x!x)$', ls=':')
plt.plot(x, e, c='yellow', label='$O(x^2)$', ls='--')
plt.plot(x, f, c='orange', label='$O(2^n)$', ls=':')
plt.plot(x, g, c='brown', label='$O(n!)$', ls='--')
plt.plot(x, h, c='purple', label='$O(n^n)$', ls=':')

plt.axis([0,9,0,90])
plt.legend()
```
