Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.



```python
a = np.array(range(0, 6))
#for a in a:    
#    b = math.sqrt(a)
b = [a for a in a]
c = a ** 3
d = 2 ** a
e = 1 / (a + 1)

plt.title('Comparision of values')
plt.plot(a, b, c='green', label='Square root of a', ls='--')
plt.plot(a, c, c='red', label='a to the power of 3', ls='--')
plt.plot(a, d, c='blue', label='2 to the power of a', ls='--')
plt.plot(a, e, c='purple', label='1 divided by a plus 1', ls='--')
plt.axis([0,10, 0, 12])
plt.xlabel('The value of a')
plt.ylabel('The value of the equation')
plt.legend()
```
