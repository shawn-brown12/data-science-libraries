Class examples

```python
#generate x values

x1 = range(0, 100)

#calculate the y2 value
y1 = [math.sqrt(n) for n in x1]

#plot the data

plt.plot(x1, y1)

#customize the chart

plt.title('$ y = \sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('sqrt{x}$')
plt.xlim(0, 100)
plt.ylim(0, 10)

#grid lines

plt.grid(True, ls='--')
```

```python
#cube of x

#generate x2 values

x2 = range(-100, 100)

#calculate the y2 value
y2 = [n ** 3 for n in x2]

#plot the data

plt.plot(x2, y2)

#customize the chart

plt.title('$ y = x^3$')
plt.xlabel('$x$')
plt.ylabel('x^3$')
plt.xlim(0, 10)
plt.ylim(0, 40)

#grid lines

plt.grid(True, ls='--')
```

```python
# 2 ^ x

#generate x2 values
x3 = [x * 0.1 for x in range(0, 150)]

#calculate the y2 value
y3 = [2 ** n for n in x3]

#plot the data

plt.plot(x3, y3)

#customize the chart

plt.title('$ y = 2^x$')
plt.xlabel('$x$')
plt.ylabel('2^x$')
plt.xlim(0, 5)
plt.ylim(0, 50)

#grid lines

plt.grid(True, ls='--')
```

```python
#generate x2 values
x4 = [x * 0.1 for x in range(0, 150)]

#calculate the y2 value
y4 = [1/(n+1) for n in x4]

#plot the data
plt.plot(x4, y4)

#customize the chart
plt.title(r'$ y = \frac{1}{x+1}$')
plt.xlabel('$x$')
plt.ylabel(r'$\frac{1}{x+1}$')
plt.xlim(0, 14)
plt.ylim(0, 2)

#grid lines
plt.grid(True, ls='--')
```

Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.



```python
a = np.array(range(0, 6))
#for a in a:    
#    b = math.sqrt(a)
b = [a for a in a]
c = a ** 3
d = 2 ** a
e = 1 / (a + 1)

plt.title('Comparision of values')
plt.plot(a, b, c='green', label='Square root of a', ls='--')
plt.plot(a, c, c='red', label='a to the power of 3', ls='--')
plt.plot(a, d, c='blue', label='2 to the power of a', ls='--')
plt.plot(a, e, c='purple', label='1 divided by a plus 1', ls='--')
plt.axis([0,10, 0, 12])
plt.xlabel('The value of a')
plt.ylabel('The value of the equation')
plt.legend()
```
