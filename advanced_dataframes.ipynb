```python
import pandas as pd
import numpy as np
```

```python
# Create the users DataFrame.

users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['bob', 'joe', 'sally', 'adam', 'jane', 'mike'],
    'role_id': [1, 2, 3, 3, np.nan, np.nan]
})
users
```

```python
# Create the roles DataFrame

roles = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['admin', 'author', 'reviewer', 'commenter']
})
roles
```

What is the result of using a right join on the DataFrames?



```python
users.merge(roles,
            left_on='role_id',
            right_on='id',
            how='right',
            indicator=True
           )
```

What is the result of using an outer join on the DataFrames?



```python
users.merge(roles,
            left_on='role_id',
            right_on='id',
            how='outer',
            indicator=True
           )
```

What happens if you drop the foreign keys from the DataFrames and try to merge them?



```python
users.drop(columns='role_id')
roles.drop(columns='id')

users.merge(roles,
           how='inner',
           indicator=True)
```

Create a column named mileage_difference like you did in the DataFrames exercises; this column should contain the difference between highway and city mileage for each car.



```python
mpg['mileage_difference'] = mpg.highway - mpg.city
mpg
```

Create a column named average_mileage like you did in the DataFrames exercises; this is the mean of the city and highway mileage.



```python
mpg['avg_mileage'] = (mpg.highway + mpg.city) / 2
mpg
```

Create a new column on the mpg dataset named is_automatic that holds boolean values denoting whether the car has an automatic transmission.



```python
mpg.groupby('trans').sum()
```

```python
#can use np.where which more than one condition, like below
#format np.where((dataframe.column ==/>/whatever fits your need 'condition to meet') | 
#                 (dataframe.column ==/>/whatever 'condition to meet', if_condition met, if_not_met))

mpg['is_automatic'] = np.where((mpg.trans == 'manual(m6)') |
                               (mpg.trans == 'manual(m5)'), False, True)
                              
mpg                       
```

Using the mpg dataset, find out which which manufacturer has the best miles per gallon on average?



```python
mpg.groupby('manufacturer')['avg_mileage'].agg('max')
```

Do automatic or manual cars have better miles per gallon?



```python
mpg.groupby('is_automatic')['avg_mileage'].agg(['max', 'min', 'median', 'mean'])
```
