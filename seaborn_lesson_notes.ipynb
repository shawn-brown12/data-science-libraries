```python
tips = sns.load_dataset('tips')

tips.head()
```

```python
tips.describe()
```

```python
tips.info()
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, s=100, color='.2', marker='+')
plt.title('Total Bill vs Tip Amount')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')

plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', col='sex', data=tips)

plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', col='time', hue='size', style='smoker', data=tips)

plt.show()
```

```python
sns.histplot(tips.tip, kde=True)
plt.show()
```

```python
sns.boxplot(data=tips, x='tip')
plt.show()
```

```python
sns.boxplot(data=tips, y='tip', x='size', hue='smoker')
plt.show()
```

```python
sns.pairplot(tips)
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip')
```
